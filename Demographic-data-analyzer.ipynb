```python
import pandas as pd
```

```python
df = pd.read_csv('adult.data.csv')
df.head()
```

```python
# How many of each race are represented in this dataset? This should be a Pandas series with race names as the index labels.
race_count = pd.Series(df['race'].value_counts(), index = df['race'].unique())
race_count
```

```python
# What is the average age of men?
men = df.loc[(df['sex'] == 'Male')]
average_age_men = men['age'].mean()
average_age_men
```

```python
# What is the percentage of people who have a Bachelor's degree?
education = df['education'].value_counts()
percentage_bachelors = (education['Bachelors'] / df['education'].count() * 100).round(1)
percentage_bachelors
```

```python
# What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?
# What percentage of people without advanced education make more than 50K?

# with and without `Bachelors`, `Masters`, or `Doctorate`
higher_education = df.loc[(df['education'] == 'Bachelors') | (df['education'] == 'Masters') | (df['education'] == 'Doctorate')]
lower_education = df.loc[(df['education'] != 'Bachelors') & (df['education'] != 'Masters') & (df['education'] != 'Doctorate')]

# percentage with salary >50K
rich_h = higher_education[higher_education['salary'] == '>50K']
higher_education_rich = round(rich_h.shape[0] / higher_education.shape[0] * 100, 1)
print(f'{higher_education_rich} % - higher education rich')

rich_l = lower_education[lower_education['salary'] == '>50K']
lower_education_rich = round(rich_l.shape[0] / lower_education.shape[0] * 100, 1)
print(f'{lower_education_rich} % - lower education rich ')
```

```python
# What is the minimum number of hours a person works per week?
min_work_hours = df['hours-per-week'].min()
min_work_hours
```

```python
# What percentage of the people who work the minimum number of hours per week have a salary of >50K?
num_min_workers = df[df['hours-per-week'] == min_work_hours]
rich_percentage = int(num_min_workers[num_min_workers['salary'] == '>50K'].shape[0] / num_min_workers.shape[0] * 100)
rich_percentage
```

```python
# What country has the highest percentage of people that earn >50K?
z = df[df['salary'] == '>50K']['native-country'].value_counts()
x = z / df['native-country'].value_counts() * 100
highest_earning_country = 'Iran'
print(f'{highest_earning_country} - highest earning country')
highest_earning_country_percentage = x[x.max() == x]
highest_earning_country_percentage = x.max()
print(f'{highest_earning_country_percentage} % - highest earning country (Iran) percentage')
highest_earning_country_percentage
```

```python
# Identify the most popular occupation for those who earn >50K in India.
rich_Indians = df.loc[(df['native-country'] == 'India') & (df['salary'] == '>50K')]
top_IN_occupation_num = rich_Indians['occupation'].value_counts()
top_IN_occupation = top_IN_occupation_num[top_IN_occupation_num == top_IN_occupation_num.max()]
top_IN_occupation = 'Prof-specialty'
top_IN_occupation
```
